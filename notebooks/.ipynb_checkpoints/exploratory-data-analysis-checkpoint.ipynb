```python
df= pd.read_csv("/kaggle/input/ece-143-group-5/final_combined_dataset_v3.csv")
```

I will drop stop_id and search_deatil_id

```python
drop_cols = ['stop_id', 'search_details_id']
df = df.drop(columns = drop_cols)
df
```

```python
# Selecting relevant numerical columns for correlation heatmap
numerical_columns = df.select_dtypes(include=['number']).columns

# Creating a correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Plotting the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Selecting relevant numerical columns for boxplot
numerical_columns = df.select_dtypes(include=['number']).columns

# Plotting boxplot
plt.figure(figsize=(15, 20))
sns.boxplot(data=df[numerical_columns])
plt.title('Boxplot for Numerical Columns')
plt.show()
```

```python
# Define a mapping for search_details_type
type_mapping = {
    0: 'ActionTaken',
    1: 'ActionTakenOther',
    2: 'SearchBasis',
    3: 'SearchBasisOther',
    4: 'SearchType'
}

# Map the values in the 'search_details_type' column
df['search_details_type'] = df['search_details_type'].map(type_mapping)

# Check the distribution of stop causes
stop_cause_counts = df['search_details_type'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=stop_cause_counts.index, y=stop_cause_counts.values, palette='viridis')
plt.title('Distribution of Stop Causes')
plt.xlabel('Stop Cause')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(x='service_area', hue='stop_cause', data=df, palette='Set2')
plt.title('Distribution of Stop Causes by Service Area')
plt.xlabel('Service Area')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Stop Cause', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
top_stop_causes = stop_cause_counts.head(5)  # Adjust the number of top causes as needed
plt.figure(figsize=(8, 8))
plt.pie(top_stop_causes, labels=top_stop_causes.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Top Stop Causes')
plt.show()
```

```python
# Create age bins
age_bins = [0, 18, 25, 35, 45, 55, 65, 75, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66-75', '76+']
```

```python
arrest_counts = df['arrested'].value_counts()

# Visualize the distribution of arrests
plt.figure(figsize=(6, 6))
plt.pie(arrest_counts, labels=arrest_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Distribution of Arrests')
plt.show()

# Analyze the relationship between stop causes and the likelihood of arrest
plt.figure(figsize=(12, 6))
sns.countplot(x='stop_cause', hue='arrested', data=df, palette='viridis')
plt.title('Stop Causes Distribution by Arrest Status')
plt.xlabel('Stop Causes')
plt.ylabel('Count')
plt.legend(title='Arrested', loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.show()

# Visualize the percentage of stops resulting in arrests
arrest_percentage = df['arrested'].value_counts(normalize=True) * 100

plt.figure(figsize=(6, 6))
plt.pie(arrest_percentage, labels=arrest_percentage.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Percentage of Stops Resulting in Arrests')
plt.show()
```
